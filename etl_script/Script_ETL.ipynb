```python
# 📥 1. IMPORT LIBRARIES
import pandas as pd
from sqlalchemy import create_engine
import psycopg2

# 👉 CUSTOMER
dim_customer = df3[['Customer ID', 'Customer Name', 'Segment']].drop_duplicates().reset_index(drop=True)
dim_customer['customer_key'] = dim_customer.index + 1

# 👉 PRODUCT
dim_product = df3[['Product ID', 'Product Name', 'Category', 'Sub-Category']].drop_duplicates().reset_index(drop=True)
dim_product['product_key'] = dim_product.index + 1

# 👉 ORDER
dim_order = df3[['Order ID', 'Order Date', 'Ship Date', 'Ship Mode', 'Order Priority']].drop_duplicates().reset_index(drop=True)
dim_order['order_key'] = dim_order.index + 1

# 👉 TIME
dim_time = df3[['Order Date']].drop_duplicates().reset_index(drop=True)
dim_time['time_key'] = dim_time.index + 1
dim_time['day'] = dim_time['Order Date'].dt.day
dim_time['month'] = dim_time['Order Date'].dt.month
dim_time['month_name'] = dim_time['Order Date'].dt.month_name()
dim_time['quarter'] = dim_time['Order Date'].dt.quarter
dim_time['year'] = dim_time['Order Date'].dt.year
dim_time['day_of_week'] = dim_time['Order Date'].dt.dayofweek
dim_time['is_weekend'] = dim_time['day_of_week'].isin([5, 6])

# 👉 REGION
dim_region = df3[['City', 'State', 'Country', 'Region', 'Market', 'lat', 'lng']].drop_duplicates().reset_index(drop=True)
dim_region['region_key'] = dim_region.index + 1

# 🔁 4. FOREIGN KEY LOOKUPS
customer_lookup = dim_customer.set_index('Customer ID')['customer_key'].to_dict()
product_lookup = dim_product.set_index('Product ID')['product_key'].to_dict()
order_lookup = dim_order.set_index('Order ID')['order_key'].to_dict()
time_lookup = dim_time.set_index('Order Date')['time_key'].to_dict()
region_lookup = dim_region.set_index(['City', 'State', 'Country', 'Region', 'Market', 'lat', 'lng'])['region_key'].to_dict()

# 🧾 5. CREATE FACT TABLE
fact_sales = df3[['Order ID', 'Product ID', 'Customer ID', 'Order Date', 'Quantity', 'Sales', 'Discount', 'Profit', 'Shipping Cost',
                  'City', 'State', 'Country', 'Region', 'Market', 'lat', 'lng']].copy()

# 🔗 Map foreign keys
fact_sales['customer_key'] = fact_sales['Customer ID'].map(customer_lookup)
fact_sales['product_key'] = fact_sales['Product ID'].map(product_lookup)
fact_sales['order_key'] = fact_sales['Order ID'].map(order_lookup)
fact_sales['time_key'] = fact_sales['Order Date'].map(time_lookup)

fact_sales['region_key'] = fact_sales.set_index(['City', 'State', 'Country', 'Region', 'Market', 'lat', 'lng']).index.map(region_lookup)

# 🧼 Clean & reorder
fact_sales = fact_sales[[
    'order_key',
    'product_key',
    'customer_key',
    'region_key',
    'time_key',
    'Quantity',
    'Sales',
    'Discount',
    'Profit',
    'Shipping Cost'
]]

# 🐍 6. CONVERT TO SNAKE_CASE
def snake_case(df3):
    df3.columns = [col.lower().replace(' ', '_') for col in df3.columns]
    return df3

dim_customer = snake_case(dim_customer)
dim_product = snake_case(dim_product)
dim_order = snake_case(dim_order)
dim_time = snake_case(dim_time)
dim_region = snake_case(dim_region)
fact_sales = snake_case(fact_sales)

# 🛠️ 7. CONNECT TO POSTGRESQL
DB_USER = 'postgres'
DB_PASS = '18agustuz203'
DB_HOST = 'localhost'
DB_PORT = '5432'
DB_NAME = 'dwh_superstore2'

engine = create_engine(f'postgresql+psycopg2://{DB_USER}:{DB_PASS}@{DB_HOST}:{DB_PORT}/{DB_NAME}')

# 💾 8. PUSH ALL TO DATABASE
dim_customer.to_sql('dim_customer', engine, if_exists='replace', index=False)
dim_product.to_sql('dim_product', engine, if_exists='replace', index=False)
dim_order.to_sql('dim_order', engine, if_exists='replace', index=False)
dim_time.to_sql('dim_time', engine, if_exists='replace', index=False)
dim_region.to_sql('dim_region', engine, if_exists='replace', index=False)
fact_sales.to_sql('fact_sales', engine, if_exists='replace', index=False)

print("✅ ETL completed — all tables including `region_key` successfully pushed to PostgreSQL!")
```
